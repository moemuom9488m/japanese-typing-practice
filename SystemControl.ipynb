# 🎮 日文打字系統 控制台
使用此 Notebook 來管理你的 Docker 容器。請確保你已安裝 [Docker Desktop](https://www.docker.com/products/docker-desktop/)。

## 1. 🚀 啟動系統 (Start)
啟動後可訪問 [http://localhost:5173](http://localhost:5173)

```python
!docker compose up -d --build
```

## 2. 📋 查看運行狀態與日誌 (Monitor)

```python
!docker compose ps
!docker compose logs --tail=20
```

## 3. 🛠️ 同步套件至本地 (Maintenance)
當編輯器（如 VS Code）沒有提示時執行此格，將 `node_modules` 下載至主機。

```python
!docker run --rm -v ${PWD}:/app -w /app node:20 npm install
```

## 4. 🛑 關閉系統 (Stop)
停止服務並清理容器。

```python
!docker compose down
```

## 🌐 外部 IP 存取提示
若要在區域網路內使用，請點擊下方儲存格取得你的本機 IP：

```python
!ipconfig | findstr /i "IPv4"
```
